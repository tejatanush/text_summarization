# **Model Description:**

Developed a model for Text summarization using transformers pipeline. This model takes input of an article and summarizes the article in to less length which is readable with understandable words.

# 1.Import Libraries

```python
from transformers import pipeline
summarizer=pipeline('summarization')
```

# 2. Build summarization model

```python
def summarized_text(article):
  words=article.split()
  word_count=len(words)
  return summarizer(article,max_length=round(word_count*0.6),min_length=100,do_sample=False)
```

We set our model to summarize with a minimum length of 100 words and maximum length of (no.of words in article*0.6)

#3. Take input and display summarized text.

```python
for i in range(1,4):
  print(i)
  article=input()
  print(summarized_text(article))
```

We can see our model displayed summarized text with the given inputs with desired features and understandable words.
